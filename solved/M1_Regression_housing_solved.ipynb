# Ejercicio de regresión

Empezamos especificando las librerías que vamos a usar

```python
from sklearn.datasets import fetch_california_housing
from sklearn import datasets, linear_model, utils, preprocessing
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
```

Cargamos el conjunto de datos y realizamos una exploración inicial:

```python
housing = fetch_california_housing(as_frame=True)

print(housing.keys())
print("Target: ", housing.target_names)
df = housing.frame

df.head()
```

```python
df.describe()
```

Qué podemos hacer con una sola característica?

```python
X = df.drop(['MedHouseVal'], axis=1)
y = df['MedHouseVal']

X1 = df[['MedInc']]

X1.head()
```

```python
linreg = linear_model.LinearRegression(fit_intercept=True)  # This is the default
linreg.fit(X1.values, y)
plt.plot(X1, y, '.')
x = np.linspace(0, 10).reshape(-1,1)
plt.plot(x, linreg.predict(x), '-')
plt.xlabel('AveBedrms')
plt.ylabel('Home Price')
plt.show()
```
